```python
df_train = pd.read_csv("C:\\Users\\velja\\OneDrive\\Desktop\\takmicenje\\isurance_vezba\\data\\train.csv")
df_test = pd.read_csv("C:\\Users\\velja\\OneDrive\\Desktop\\takmicenje\\isurance_vezba\\data\\test.csv")

df_train = clean_a(df_train)
df_test = clean_a(df_test)

print(df_train)
X_train = df_train.drop(columns=["Premium"])
X_train = X_train.select_dtypes(include='number')
X_train = X_train.drop(columns=["ID", "Distribution_channel", "Seniority", "Policies_in_force", "Max_policies", "Max_products", "Lapse" 
                                ,"Cost_claims_year", "N_claims_year", "Area"])

y_train = df_train["Premium"]

X_test = df_test.drop(columns=["Premium"])
X_test = X_test.select_dtypes(include='number')
X_test = X_test.drop(columns=["ID", "Distribution_channel", "Seniority", "Policies_in_force", "Max_policies", "Max_products", "Lapse" 
                                ,"Cost_claims_year", "N_claims_year", "Area"])

y_test = df_test["Premium"]

```

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, r2_score, make_scorer
from sklearn.model_selection import RandomizedSearchCV


xgb = XGBRegressor(random_state=42, n_jobs=-1)

param_distributions = {
    'n_estimators': [100, 200, 300, 400],
    'learning_rate': [0.01, 0.03, 0.05, 0.07, 0.1],
    'max_depth': [4, 5, 6, 7, 8],
    'subsample': [0.7, 0.8, 0.9, 1.0],
    'colsample_bytree': [0.7, 0.8, 0.9, 1.0],
    'gamma': [0, 0.1, 0.2],
    'reg_alpha': [0, 0.01, 0.1],
    'reg_lambda': [1, 1.5, 2]
}

mae_scorer = make_scorer(mean_absolute_error, greater_is_better=False)

random_search = RandomizedSearchCV(
    estimator=xgb,
    param_distributions=param_distributions,
    n_iter=100,
    scoring=mae_scorer,
    cv=5,
    verbose=2,
    random_state=42,
    n_jobs=-1
)

random_search.fit(X_train, y_train)

print("Best parameters found:", random_search.best_params_)

best_model = random_search.best_estimator_
y_pred = best_model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
print("MAE:", mae)

r2 = r2_score(y_test, y_pred)
print("R^2:", r2)

print("Mean of y_test:", y_test.mean())
```
